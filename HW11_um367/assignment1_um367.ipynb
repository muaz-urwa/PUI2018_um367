# 1: event detection 

## Event detection: Identify the most prominent event. There is a very significant drop (>3-sigma) in all time series. Find the date and see what  may have caused this event 



`````python
print('Number of Time series: ',mta.shape[0]*mta.shape[1])
`````

`````python
stdMatrix = np.nanstd(mta,axis=2)
broadcastedStdMatrix = np.repeat(stdMatrix[:, :, np.newaxis], mta.shape[2], axis=2)

meanMatrix = np.nanmean(mta,axis=2)
broadcastedMeanMatrix = np.repeat(meanMatrix[:, :, np.newaxis], mta.shape[2], axis=2)

#zscore matrix
zmta = np.divide((mta-broadcastedMeanMatrix),broadcastedStdMatrix) 

# < 3 sigma matrix
zmta3 = zmta <= -3

collapsedzmta3 = np.nansum(zmta3,axis=(0,1))
collapsedzmta3.shape
`````

`````python
pl.figure(figsize=(25,10))  
pl.bar(np.array(range(collapsedzmta3.size))+1,collapsedzmta3.reshape(collapsedzmta3.size))
pl.xlabel('weeks since 05/21/2010', fontsize=24)
pl.ylabel('Count of Timeseries with value < mean - 3std', fontsize=24)
pl.title('Histogram of anomaly counts over weeks')
`````

### This figure is a histogram which shows the count of station which were more than 3 std below the mean on that given week. It is obvious that there is an anomaly on 127 th week.

`````python
print('Max number of negative anomalies in a week: ',np.max(collapsedzmta3))
print('Respective week: ',np.argmax(collapsedzmta3)+1)
`````

`````python
#https://stackoverflow.com/questions/5189777/python-adding-3-weeks-to-any-date
import datetime
u = datetime.datetime.strptime("2010-05-21","%Y-%m-%d")
d = datetime.timedelta(weeks= np.argmax(collapsedzmta3)+1)
t = u + d
# source: https://www.onthisday.com/date/2012/october/26
print('Day of anomaly: ',t.date(),"\nHurricane Sandy makes landfall in the Bahamas killing 2 people and causing over $300 million in damage")
`````

 # 2: finding trends

## NOTE ON THE TREND DETECTION:
### you should find the *ride type* that has the steepest increase, and the steepest decrease. Notice: we are looking for something that has a more or less constant trend

### to do this you can sum or average (a weighted average is probably best) over all stations, leaving you with 23 time series, one per ridership type. 

### you can use a rolling mean for smoothing (pd.rolling_mean() for example), and you can calculate the ratio of the *number of rides* near the beginning to the *number of rides* near the end of the time series.

### notice that you should inspect the time series, algorithmically or at least by eye, to ensure you see a trend, and not just changing structure. Below is an example of a lightcurve with structure (top) and a lightcurve with a constant trend (bottom) both showing a large deviation between the beginning and the end of the time series


`````python
meanCollapse_ridertype = np.nanmean(mta, axis=0)
print(meanCollapse_ridertype.shape)
`````

`````python
for i  in range(len(ridetype)):
    #print("Ride Type: ",ridetype[i])
    first4WeekAvg = np.nansum(meanCollapse_ridertype[i][:4])
    last4WeekAvg = np.nansum(meanCollapse_ridertype[i][-4:])
    change = (last4WeekAvg - first4WeekAvg) / first4WeekAvg
    std = np.nanstd(meanCollapse_ridertype[i])
    pl.figure(figsize=(10,5))
    pl.plot(pd.Series(meanCollapse_ridertype[i]).rolling(25).mean())
    pl.title('Ride Type: '+ridetype[i]+' std: '+str(round(std,2))+' frac change: '+str(round(change,2)))
    pl.xlabel('weeks since 05/21/2010', fontsize=10)
    pl.ylabel('Rolling mean (window=25)', fontsize=24)
`````

### This figure shows the rolling means of the cumulative time series of all the rider types with window length of 25.
#### Rider type sen has the steepest positive constant slope od 2.2
#### Rider type month has steepest negative constant slope of -0.4

`````python
def getRegCoeff(Y):
    if Y.std() > 0:
        X = sm.add_constant(np.array(range(len(Y))))
        mod = sm.OLS(Y,X).fit()
        return mod.params[0],mod.params[1]
    else : 
        return (np.nan,np.nan)
`````

`````python
from sklearn.model_selection import RandomizedSearchCV


#https://towardsdatascience.com/hyperparameter-tuning-the-random-forest-in-python-using-scikit-learn-28d2aa77dd74

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 20, stop = 100, num = 3)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10,100 , num = 3)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]

# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}

print(random_grid)
`````

`````python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = rfc()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 20, cv = 3, verbose=2, random_state=42, n_jobs = -1)

# Fit the random search model
rf_random.fit(X_train, y_train)
`````

`````python
rf_random.best_params_
`````

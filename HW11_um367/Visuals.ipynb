```python
### This function downloads the pdf file, moves it into the PUIdata and reads it into a dataframe
def getPdfUrl(url,pdfFile):
    #csvFile = 'file.csv'
    urlretrieve(url, pdfFile)
    os.system("mv " + pdfFile + ' ' + os.getenv("PUIDATA"))
    #csvFilePath = os.getenv('PUIDATA') + '/' + csvFile
    #return pd.read_csv(csvFilePath)
```

```python
## https://www.geeksforgeeks.org/working-with-pdf-files-in-python/

def extractText(pdfFilePath):
    # creating a pdf file object 
    pdfFileObj = open(pdfFilePath, 'rb') 

    # creating a pdf reader object 
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj) 
  
    # printing number of pages in pdf file 
    n = pdfReader.numPages

    text = ""
    for i in range(n):
        # creating a page object 
        pageObj = pdfReader.getPage(i)
    
        # extracting text from page 
        text += pageObj.extractText()
 
    # closing the pdf file object 
    pdfFileObj.close() 
    
    return text
```

```python
# load doc into memory
def load_doc(filename):
    # open the file as read only
    file = open(filename, 'r')
    # read all text
    text = file.read()
    # close the file
    file.close()
    return text
```

```python
import string
from string import punctuation

def removePunctuation(s):
    exclude = set(string.punctuation)
    s = ''.join(ch for ch in s if ch not in exclude)
    return s
```

```python
import json
file = 'sage_research_methods.json'
text = load_doc(file)
loaded_json = json.loads(text)
data = loaded_json['@graph']

concepts = []
for i in range(len(data)):
    #print(i)
    if 'skos:prefLabel' in data[i]:
        value = data[i]['skos:prefLabel']['@value']
        concepts.append(value)
concepts = [word.lower() for word in concepts]
concepts = [removePunctuation(word) for word in concepts]
```

```python
def getWordDict():
    d = {}
    for w in vocab:
        d[w]=0
    return d
```

```python
listFeatures = []

for i in range(len(texts)):
    d = getWordDict()
    tok = word_tokenize(texts[i])
    w = [word for word in tok if word.lower() not in stop_words and word.isalpha()]
    freq = nltk.FreqDist(w)
    for k in d.keys():
        #print(k)
        #print(freq[k])
        d[k] = freq[k]
    d['Lecture'] = 'UI-'+str(i)
    listFeatures.append(d)
```

```python
API_URL = 'https://fb55.carto.com:443/api/v2/sql?q='

def queryCartoDB(query, formatting = 'CSV', source = API_URL):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    
    data = urlencode({'format': formatting, 'q': query}).encode("utf-8")
    try:
        response = urlopen(source, data)
        return response.read()
    except HTTPError as e:
        raise (ValueError('\n'.join(ast.literal_eval(e.readline())['error'])))
        
def get_data(query):
    try:
        return pd.read_csv(io(queryCartoDB(query)), sep = ',')
    except ValueError as v:
        print (str(v))
```

# Task 1

#### Sort data by start_station_id, tripduration 
  ####  -Only checking trips with duration <= 3 hours 

```python
query = '''
SELECT *
FROM citibike
WHERE tripduration <= 3*3600
Order BY start_station_id, tripduration
'''
get_data(query)
```

### Only show the top/last 10 records

```python
query = '''
SELECT *
FROM citibike
limit 10
'''
get_data(query)
```

#### List all unique start_station_id values

```python
query = '''
SELECT Distinct start_station_id
FROM citibike
'''
## Removing a redundent column by column selection
get_data(query)[['start_station_id']]
```

### Count the number of trips

```python
query = '''
SELECT Count(*)
FROM citibike
'''
get_data(query)
```

#### Find the average/min/max trip duration

```python
query = '''
SELECT Avg(tripduration), Max(tripduration), Min(tripduration)
FROM citibike
'''
get_data(query)
```

# Task 2

#### Selecting trips started on Feb-02-2015 only (since no records exist on this date i use 2015-02-01 instead)

```python
query = '''
SELECT *
FROM citibike
WHERE starttime::date = '2015-02-01'
'''

get_data(query)
```

### Selecting trips started on the weekends

```python
query = '''
SELECT *
FROM citibike
WHERE EXTRACT(DOW FROM starttime::TIMESTAMP) in (0,6)
'''

get_data(query)
```

### Apparently there is only one date in the data. Lets see it using query

```python
query = '''
SELECT Distinct starttime::date AS date, EXTRACT(DOW FROM starttime::TIMESTAMP) AS DOW
FROM citibike
'''

get_data(query)
```

#### There is only one date and the corresponding day of the week is sunday.

### What are average trip duration during weekends?

```python
query = '''
SELECT Avg(tripduration) AS AvgTripDuration
FROM citibike
WHERE EXTRACT(DOW FROM starttime::TIMESTAMP) in (0,6)
'''
get_data(query)[['avgtripduration']]
```

#### Can we do the same for weekday?

```python
query = '''
SELECT Avg(tripduration) AS AvgTripDuration
FROM citibike
WHERE EXTRACT(DOW FROM starttime::TIMESTAMP) in (1,2,3,4,5)
'''
get_data(query)[['avgtripduration']]
```

#### Showing the list of start station locations 

```python
query = '''
SELECT Distinct start_station_name
FROM citibike
'''
get_data(query)[['start_station_name']]
```

#### Showing the number of trips started per station 

```python
query = '''
SELECT start_station_name, Count(*)
FROM citibike
Group By start_station_name
'''
get_data(query)
```

#### Trips per station but only for stations within 500m of Time Square! The coordinates of Time Square is (40.7577,-73.9857)

```python
query = '''
Select start_station_name, Count(*)
From citibike
WHERE ST_DWithin(ST_MakePoint(start_station_latitude,start_station_longitude)::geography,
                 ST_GeogFromText('POINT(40.7577 -73.9857)'),
                 500, false)
Group By start_station_name
'''

get_data(query)
```

#### Find the station that had the longest average trip duration during weekends and within 500m of TimeSquare! 

```python
query = '''
Select start_station_name, Avg(tripduration)
From citibike
WHERE ST_DWithin(ST_MakePoint(start_station_latitude,start_station_longitude)::geography,
                 ST_GeogFromText('POINT(40.7577 -73.9857)'),
                 500, false)
       AND
      EXTRACT(DOW FROM starttime::TIMESTAMP) in (0,6)               
Group By start_station_name
Order BY Avg(tripduration) Desc
Limit 1
'''

get_data(query)
```

#### Extra: create lines for trips started from stations within 500m of Times Squares and lasted less than 2 hours. The number of trips per each pair of stations are output as attributes of these lines. 

```python
query = '''
Select ST_MakeLine(ST_MakePoint(start_station_latitude,start_station_longitude),ST_MakePoint(end_station_latitude,end_station_longitude)), COunt(*)
From citibike
WHERE ST_DWithin(ST_MakePoint(start_station_latitude,start_station_longitude)::geography,
                 ST_GeogFromText('POINT(40.7577 -73.9857)'),
                 500, false)   
                 AND
      tripduration < 2*3600
Group By ST_MakeLine(ST_MakePoint(start_station_latitude,start_station_longitude),ST_MakePoint(end_station_latitude,end_station_longitude))
'''

get_data(query)
```

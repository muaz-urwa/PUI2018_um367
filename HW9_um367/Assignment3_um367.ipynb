```python
def get_google_sheet(spreadsheet_id, range_name):
    # The file token.json stores the user's access and refresh tokens, and is
    # created automatically when the authorization flow completes for the first
    # time.
    store = file.Storage('token.json')
    creds = store.get()
    if not creds or creds.invalid:
        flow = client.flow_from_clientsecrets('credentials.json', SCOPES)
        creds = tools.run_flow(flow, store)
    service = build('sheets', 'v4', http=creds.authorize(Http()))

    
    # Call the Sheets API
    sheet = service.spreadsheets()
    result = sheet.values().get(spreadsheetId=SAMPLE_SPREADSHEET_ID,
                                range=SAMPLE_RANGE_NAME).execute()
    return result
```

```python
### This function downloads the csv file, moves it into the PUIdata and reads it into a dataframe
def getDataFrameFromCsvUrl(url,csvFile):
    #csvFile = 'file.csv'
    urlretrieve(url, csvFile)
    os.system("mv " + csvFile + ' ' + os.getenv("PUIDATA"))
    csvFilePath = os.getenv('PUIDATA') + '/' + csvFile
    return pd.read_csv(csvFilePath)
```

```python
### This function downloads the zip file, unzips it into the PUIdata and reads it into a gepandas dataframe
def getDataFrameFromFileZipUrl(url,filename):
    filenames = ! ls $PUIDATA 
    urlretrieve(url, filename+'.gz')
    os.system('unzip -d $PUIDATA '+filename+'.gz')
    os.system('rm '+filename+'.gz')
    filenames2 = ! ls $PUIDATA 
    shapeFile = list(set(filenames2).difference(filenames))[0]
    shapeFilePath = os.getenv('PUIDATA') + '/' + shapeFile
    return pd.read_csv(shapeFilePath)
```

```python
df['shootingspp'] = df['Number of mass shootings'] / df['pop'] * 1e7
df['armspp'] = df['Average total all civilian firearms'] / df['pop'] * 10

new_df = df[['shootingspp','armspp']]
```

```python
new_df.head(3)
```

```python
# OLS

olsmod = smf.ols(formula='shootingspp ~ armspp', data = new_df).fit()
olsmod.summary()
```

```python
# WLS
wlsmod = smf.WLS(new_df.shootingspp ,sm.add_constant(new_df.armspp), weights=np.sqrt(new_df.armspp)).fit()
```

```python
poly = np.poly1d(np.polyfit(new_df.armspp, new_df.shootingspp, 1))
slope=poly[0]
intercept = poly[1]
```

```python
import seaborn as sn
from statsmodels.graphics.api import abline_plot
```

```python
figure, ax = plt.subplots(1,figsize=(14, 15))

sn.regplot(x=new_df.armspp, y=new_df.shootingspp, color="g",ax=ax, label='seaborn')
abline_plot(model_results=olsmod, ax=ax, color='red',label='OLS')
ax.plot(new_df.armspp, slope*new_df.armspp + intercept,color='blue',label='ployfit')
ax.plot(new_df.armspp,wlsmod.fittedvalues,label='wls')
ax.errorbar(new_df.armspp, new_df.shootingspp, 
            yerr = np.sqrt(new_df.armspp), fmt = '.',label='data')
ax.legend()
```
